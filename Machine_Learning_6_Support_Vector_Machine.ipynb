```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"E:\Mis Carpetas\Documentos\Data Science\social_net_ads.csv")
```

```python
df.head(5)
```

```python
X = df[["Age","EstimatedSalary"]]
y = df["Purchased"]
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
from sklearn.svm import SVC

svc = SVC(C=1,kernel="linear" ,random_state=0)
svc.fit(X_train, y_train)
```

```python
y_train_pred = svc.predict(X_train)
y_test_pred = svc.predict(X_test)
```

```python
svc.score(X_train, y_train)
```

```python
svc.score(X_test, y_test)
```

```python
from sklearn.metrics import f1_score

f1_score(y_train, y_train_pred)
```

```python
f1_score(y_test, y_test_pred)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import plot_confusion_matrix
```

```python
confusion_matrix(y_test, y_test_pred)
```

```python
plot_confusion_matrix(svc, X_test, y_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, y_test_pred))
```

```python
from matplotlib.colors import ListedColormap


def plot_decision_boundaries(X, Y):
  X_Set, Y_Set = X, Y
  X1, X2 = np.meshgrid(np.arange(start = X_Set[:, 0].min() - 1, stop = X_Set[:, 0].max() + 1, step = 0.01),
                      np.arange(start = X_Set[:, 1].min() - 1, stop = X_Set[:, 1].max() + 1, step = 0.01))

  plt.figure(figsize=(10,5))
  plt.contourf(X1, X2, svc.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
              alpha = 0.75, cmap = ListedColormap(('red', 'green')))
  plt.xlim(X1.min(), X1.max())
  plt.ylim(X2.min(), X2.max())
  for i, j in enumerate(np.unique(Y_Set)):
      plt.scatter(X_Set[Y_Set == j, 0], X_Set[Y_Set == j, 1],
                  c = ListedColormap(('red', 'green'))(i), label = j)
  plt.title('Support Vector Machine', fontsize=20)
  plt.xlabel('Edad', fontsize=15)
  plt.ylabel('Salario Estimado', fontsize=15)
  plt.legend()
  plt.show()

plot_decision_boundaries(X_train, y_train)
```

```python
plot_decision_boundaries(X_test, y_test)
```

```python
svc = SVC(C=1, kernel="rbf", random_state=0)

svc.fit(X_train, y_train)
y_test_pred = svc.predict(X_test)

plot_decision_boundaries(X_train, y_train)
```

```python
plot_decision_boundaries(X_test, y_test)
```

```python
svc = SVC(C=0.01, kernel="linear", random_state=0)

svc.fit(X_train, y_train)
y_test_pred = svc.predict(X_test)

plot_decision_boundaries(X_train, y_train)
```

```python
plot_decision_boundaries(X_test, y_test)
```
